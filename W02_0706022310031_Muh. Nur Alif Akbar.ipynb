```python
# 1. Load the dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Inspect the dataset

url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv "
df = pd.read_csv(url, sep=',')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Calculate null counts and percentages
null_info = pd.DataFrame({
    'Null Count': df.isnull().sum(),
    'Null Percentage': (df.isnull().sum() / len(df)) * 100
})

# Filter out columns with no null values
null_info = null_info[null_info['Null Count'] > 0].sort_values(by='Null Count', ascending=False)

if not null_info.empty:
    print("Informasi Nilai Null per Kolom:")
    display(null_info)
else:
    print("Tidak ada nilai null dalam DataFrame.")
```

Dari sini kita bisa melihat bahwa ada beberapa nilai null yang jumlah persennya sangat besar bahkan hingga diatas 40%. Untuk sekarang aku ingin menghapus atau drop rows kolom yang memiliki jumlah null nya dibawah 1 persen saja.

```python
# Identify columns with less than 1% null values
cols_to_drop_rows = null_info[null_info['Null Percentage'] < 1].index.tolist()

# Drop rows with nulls in the identified columns
df_cleaned = df.dropna(subset=cols_to_drop_rows)

print(f"Original number of rows: {len(df)}")
print(f"Number of rows after dropping: {len(df_cleaned)}")

# Display null information for the cleaned DataFrame to verify
null_info_cleaned = pd.DataFrame({
    'Null Count': df_cleaned.isnull().sum(),
    'Null Percentage': (df_cleaned.isnull().sum() / len(df_cleaned)) * 100
})

null_info_cleaned = null_info_cleaned[null_info_cleaned['Null Count'] > 0].sort_values(by='Null Count', ascending=False)

if not null_info_cleaned.empty:
    print("\nInformasi Nilai Null per Kolom setelah dropping:")
    display(null_info_cleaned)
else:
    print("\nTidak ada nilai null dalam DataFrame setelah dropping.")
```

Untuk sekarang kita lihat satu persatu data tersebut baru kita mulai menentukan harus kita apakan data ini supaya tidak memiliki nilai null atau kita ganti null tersebut dengan something.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the value counts of the 'certificate' column
certificate_counts = df['certificate'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=certificate_counts.index, y=certificate_counts.values)
plt.title('Distribution of Certificate Types')
plt.xlabel('Certificate Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Disini data yang dimiliki certificate terlalu jomplang dengan yang lain maka kita drop saja kolom certificate karena awalnya kita drop rows yang isinya selain shm-sertifikat hak milik. Dikarenakan kolom ini hanya memiliki satu jenis data yaitu shm - certificate maka kita hapus saja.

```python
# Drop the 'certificate' column
df_cleaned = df_cleaned.drop('certificate', axis=1)

# Display the updated dataframe info to confirm the column is dropped
df_cleaned.info()
```

```python
# Calculate null counts and percentages for the cleaned DataFrame
null_info_cleaned = pd.DataFrame({
    'Null Count': df_cleaned.isnull().sum(),
    'Null Percentage': (df_cleaned.isnull().sum() / len(df_cleaned)) * 100
})

# Filter out columns with no null values
null_info_cleaned = null_info_cleaned[null_info_cleaned['Null Count'] > 0].sort_values(by='Null Count', ascending=False)

if not null_info_cleaned.empty:
    print("Informasi Nilai Null per Kolom setelah dropping 'certificate':")
    display(null_info_cleaned)
else:
    print("Tidak ada nilai null dalam DataFrame setelah dropping 'certificate'.")
```

Sekarang kita lihat property conditionnya seperti apa visual datanya

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the value counts of the 'property_condition' column including nulls
property_condition_counts_with_nulls = df_cleaned['property_condition'].value_counts(dropna=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=property_condition_counts_with_nulls.index, y=property_condition_counts_with_nulls.values)
plt.title('Distribution of Property Condition (Including Nulls)')
plt.xlabel('Property Condition')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Dikarenakan data ini seperti hanya ada dua data saja yang tinggi dan data yang lainnya jomplang, maka kita ganti missing values nya menjadi modus karena hanya ada sedikit data yang datanya mirip.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cek jumlah missing values di kolom property_condition
null_count = df_cleaned['property_condition'].isnull().sum()
total_count = len(df_cleaned)
print(f"Jumlah missing values: {null_count} dari total {total_count} ({(null_count/total_count)*100:.2f}%)")

# 2. Imputasi langsung dengan modus
if null_count > 0:
    modus = df_cleaned['property_condition'].mode()[0]
    df_cleaned['property_condition'] = df_cleaned['property_condition'].fillna(modus)
    print(f"Semua missing values diisi dengan modus: '{modus}'")

# 3. Hitung kembali distribusi setelah imputasi
property_condition_counts = df_cleaned['property_condition'].value_counts()

# 4. Visualisasi distribusi
plt.figure(figsize=(10, 6))
sns.barplot(x=property_condition_counts.index, 
            y=property_condition_counts.values, 
            palette="Blues_d")
plt.title('Distribution of Property Condition (After Filling with Mode)')
plt.xlabel('Property Condition')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the value counts of the 'furnishing' column
furnishing_counts = df_cleaned['furnishing'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=furnishing_counts.index, y=furnishing_counts.values)
plt.title('Distribution of Furnishing')
plt.xlabel('Furnishing Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Sama seperti kolom sebelumnya, kolom ini pun memiliki kesamaan jika datanya terlalu jomplang, dan mengurangi bias karena mayoritas datanya memang seperti itu atau modus.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Imputasi null dengan modus
modus_furnishing = df_cleaned['furnishing'].mode()[0]
df_cleaned['furnishing'] = df_cleaned['furnishing'].fillna(modus_furnishing)

# 2. Distribusi setelah imputasi
furnishing_after = df_cleaned['furnishing'].value_counts()

# 3. Visualisasi
plt.figure(figsize=(8,6))
sns.barplot(x=furnishing_after.index.astype(str),
            y=furnishing_after.values,
            palette="Greens_d")

plt.title("Distribusi Furnishing (Sesudah Imputasi Null dengan Modus)")
plt.xlabel("Furnishing")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")

# Tambahkan label di atas bar
for i, v in enumerate(furnishing_after.values):
    plt.text(i, v + 20, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

Jika bentuk datanya seperti ini maka untuk langkah awalnya adalah hapus outlier nya terlebih dahulu lalu isi null dengan median

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot of the 'year_built' column
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned, x='year_built')
plt.title('Box Plot of Year Built (Including Outliers)')
plt.xlabel('Year Built')
plt.tight_layout()
plt.show()
```

```python
# Calculate Q1, Q3, and IQR for 'year_built'
Q1_year_built = df_cleaned['year_built'].quantile(0.25)
Q3_year_built = df_cleaned['year_built'].quantile(0.75)
IQR_year_built = Q3_year_built - Q1_year_built

# Define bounds for outliers
lower_bound_year_built = Q1_year_built - 1.5 * IQR_year_built
upper_bound_year_built = Q3_year_built + 1.5 * IQR_year_built

# Filter out the outliers
df_cleaned_no_outliers = df_cleaned[(df_cleaned['year_built'] >= lower_bound_year_built) & (df_cleaned['year_built'] <= upper_bound_year_built)]

print(f"Original number of rows: {len(df_cleaned)}")
print(f"Number of rows after removing outliers: {len(df_cleaned_no_outliers)}")
```

```python
# --- Imputasi hanya untuk year_built dengan median ---
median_year_built = df_cleaned['year_built'].median()
df_cleaned['year_built'] = df_cleaned['year_built'].fillna(median_year_built)

print(f"Missing values in 'year_built' filled with median: {median_year_built}")

# --- Cek apakah masih ada null ---
null_info_after_year_built = pd.DataFrame({
    'Null Count': df_cleaned.isnull().sum(),
    'Null Percentage': (df_cleaned.isnull().sum() / len(df_cleaned)) * 100
})
null_info_after_year_built = null_info_after_year_built[
    null_info_after_year_built['Null Count'] > 0
].sort_values(by='Null Count', ascending=False)


```

Dari sini year_built telah terisi dengan median missing values nya dengan cara pada awalnya kita cari outliernya dan hapus semua outliernya lalu baru kita cari median pada year_built untuk mengisi missing values atau nilai nullnya.

No 3 Distribusi kategori umur bangunan

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi boxplot building_age
plt.figure(figsize=(8,5))
sns.boxplot(x=df_cleaned['building_age'], color="skyblue")
plt.title("Boxplot - Building Age (Umur Bangunan)", fontsize=14)
plt.xlabel("Building Age (tahun)")
plt.show()

# Cek nilai outlier dengan IQR
Q1 = df_cleaned['building_age'].quantile(0.25)
Q3 = df_cleaned['building_age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_cleaned[(df_cleaned['building_age'] < lower_bound) | (df_cleaned['building_age'] > upper_bound)]

print("Jumlah outlier building_age:", len(outliers))
print("Contoh outlier:")
print(outliers[['building_age']].head())
```

```python
import pandas as pd

# Definisikan bin dan label
bins = [0, 5, 15, 30, 100]  
labels = ['Baru (0-5 th)', 'Muda (6-15 th)', 'Sedang (16-30 th)', 'Tua (30+ th)']

# Binning umur bangunan
df_cleaned['building_age_category'] = pd.cut(df_cleaned['building_age'], bins=bins, labels=labels, include_lowest=True)

# Lihat distribusi kategori
print(df_cleaned['building_age_category'].value_counts())

# Visualisasi hasil binning
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(x='building_age_category', data=df_cleaned, hue='building_age_category', palette="viridis", legend=False)
plt.title("Distribusi Kategori Umur Bangunan")
plt.xlabel("Kategori Umur Bangunan")
plt.ylabel("Jumlah Properti")
plt.show()

```

Insight: Banyak jumlah properti atau rumah yang baru saja dibangun atau sekitar umur 0-5 tahun yang mana pembangunan berjalan dengan baik dikarenakan banyak properti yang baru.
