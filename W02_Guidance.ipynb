```python
import pandas as pd

# Download the CSV file from the URL
url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/data_property_surabaya.csv"
df = pd.read_csv(url, sep=';')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Assuming 'id' is the name of the column you want to drop.
# If not, replace 'id' with the actual column name

df = df.drop('ID', axis=1)
df.head()
```

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]


# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Null Count', y=null_df.index, data=null_df, orient='h', palette='rainbow')
plt.xlabel('Number of Null Values', fontsize=12)
plt.ylabel('Columns', fontsize=12)
plt.title('Null Values per Column', fontsize=14)

# Add labels to the bars
for i, v in enumerate(null_df['Null Count']):
    plt.text(v + 1, i, str(v), color='black', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Assuming 'id' is the name of the column you want to drop.
# If not, replace 'id' with the actual column name

df = df.drop('ID', axis=1, errors='ignore')
df.head()
```

```python
# Replace values in 'Tingkat/Lantai' column
df['Tingkat/Lantai'] = df['Tingkat/Lantai'].replace({1.5: 2, 2.5: 3, 3.5: 4, " 3.": 3})
df['Tingkat/Lantai'].unique()
```

```python
# Fill NaN values with a suitable default (e.g., 0 or the mean) or drop them
# Here we fill with 0
df['Tingkat/Lantai'] = df['Tingkat/Lantai'].fillna(0)

# Convert to integer
df['Tingkat/Lantai'] = df['Tingkat/Lantai'].astype(int)


df['Tingkat/Lantai'].unique()
```

```python
df['Listed Berapa Lama'].unique()
```

```python
# Define the mapping for 'Listed Berapa Lama'
mapping = {
    '1': ['<1 tahun', '< 1 tahun', '< 1tahun', '1 bulan', '6 bln', '1 thn', '8 bln', '3 bln',
       '4 bln', '2 bln', '2 bln ', '3 bln ', '6 bln ', '7 bln ', '1 bln ', '4 bln ',
       ' 3 bln', '3 bulan', ' 6 bulan', '6 bulan', '2 bulan', '4 bulan',
       '5 bulan', '8 bulan', '5 bln', '7 bln', '1 bln', '3 bln ', '6 bln ', '7 bln ', '1 bln ', '4 bln ', ' 3 bln',
                     '3 bulan', ' 6 bulan', '6 bulan', '2 bulan', '4 bulan', '5 bulan', '8 bulan', '14 bulan',
                     '7 bulan', '3 bulan ', '10 bulan', '6', '2', '8', '3', '4', '5', '7', '2021', '8 bulan','7 bulan', '3 bulan ', '10 bulan', '6', '2', '8', '3', '4', '5', '7'],
    '2':['2020', '12 bulan', '18', '2 tahun', '2 thn', '1 tahun', '14 bulan', '15', '18', '2 tahun'],
    '3': ['3 tahun', '3 thn']
}

# Create a new column 'Listed_Category' based on the mapping
def categorize_listed_lama(value):
    for category, values in mapping.items():
        if value in values:
            return category
    return 'Unknown'  # Handle cases where the value is not in the mapping

df['Listed Berapa Lama'] = df['Listed Berapa Lama'].apply(categorize_listed_lama)

df['Listed Berapa Lama'].unique()
```

```python
# Create a new column 'Fasilitas Publik'
df['Fasilitas Publik'] = 0

# Define the columns representing public facilities
fasilitas_columns = ['Pasar', 'Sekolah', 'Faskes/\nRumah Sakit', 'Mall', 'Tempat Ibadah', 'Makam', 'Club House', 'Transportasi Umum', 'Jalan Raya Utama']

# Iterate over the rows of the DataFrame
for index, row in df.iterrows():
  yes_count = sum(1 for col in fasilitas_columns if row[col] == 'Yes')
  if 1 <= yes_count <= 3:
    df.loc[index, 'Fasilitas Publik'] = 1
  elif 4 <= yes_count <= 6:
    df.loc[index, 'Fasilitas Publik'] = 2
  elif 7 <= yes_count <= 9:
    df.loc[index, 'Fasilitas Publik'] = 3

# Display the updated DataFrame
df.head()
```

```python
# Define the columns representing public facilities
fasilitas_columns = ['Pasar', 'Sekolah', 'Faskes/\nRumah Sakit', 'Mall', 'Tempat Ibadah', 'Makam', 'Club House', 'Transportasi Umum', 'Jalan Raya Utama']

# Drop the specified columns
df = df.drop(columns=fasilitas_columns, errors='ignore')

df.head()
```
